```python
class Musician():
    
    def __init__(self,name):
        self.name = name
        print("musician class")
    
    def test1(self):
        print("test1")
        
    def test2(self):
        print("test2")
```

```python
#inheritance başlığı
```

```python
emre = Musician("Emre")
```

```python
emre.name
```

```python
emre.test1()
```

```python
class MusicianPlus(Musician):
    
    def __init__(self,name):
        Musician.__init__(self,name)
        print("musician plus")
        
    def test3(self):
        print("test3")
    
    #override
    def test1(self):
        print("test1 test1 test1")
```

```python
emre = MusicianPlus("emre")
```

```python
type (emre)
```

```python
class Banana():
    
    def __init__(self,name):
        self.name = name
        
    def info(self):
        return f"100 calories {self.name}"
    
class Apple():
    
    def __init__(self,name):
        self.name = name
        
    def info(self):
        return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana,apple]
```

```python
class Phone():
    
    def __init__(self,name,price):
        self.name = name
        self.__price = price
    
    def info(self):
        print(f"{self.name} price is: {self.__price}")
        
    def changePrice(self,price):
        self.__price = price
```

```python
iphone = Phone("iPhone 14",500)
```

```python
iphone.info()
```

```python
iphone.__price = 400
```

```python
iphone.info()
```

```python
iphone.changePrice(300)
```

```python
iphone.info()
```

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    
    @abstractmethod
    def maxSpeed(self):
        pass
```

```python
class Tesla(Car):
    
    def maxSpeed(self):
        print("200km")
```

```python
tesla = Tesla()
tesla.maxSpeed()
```

```python
class Mercedes(Car):
    
    def maxSpeed(self):
        print("250km")
mercedes = Mercedes()
mercedes.maxSpeed()
```

```python
class Fruit():
    
    def __init__(self,name,calories):
        self.name = name
        self.calories = calories
    
    def __str__(self):
        return f"{self.name}: {self.calories} calories"
    
    def __len__(self):
        return self.calories
myFruit = Fruit("banana",150)
myFruit.calories
```

```python
myFruit.name
```

```python
class Train():
    
    def __init__(self,name):
        self.name = name
        
    def __getitem__(self,key):
        if key == "a":
            return self.name
        else:
            return "Not found"
myTrain = Train("myTrain")
myTrain["a"]
```
